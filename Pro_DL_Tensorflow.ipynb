```
inp1 = tf.placeholder(dtype = tf.float32, shape = (1,2))
inp2 = tf.placeholder(dtype =tf.float32, shape=(2,1))
ip = tf.matmul(inp1,inp2)

with tf.Session() as sess:
  i1 = np.random.normal(size = (1,2))
  i2 = np.random.normal(size = (2,1))
  inner_prod = sess.run(ip, feed_dict={inp1:i1, inp2:i2})
  print(i1,'X', i2,'=',inner_prod)
  
```

```
x_ = tf.placeholder(tf.float32, shape=(4,2), name='x-input')
y_ = tf.placeholder(tf.float32, shape=(4,1), name='y-input')
```

```

w1 = tf.Variable(tf.random_normal((2,2)))
b1 = tf.Variable(tf.zeros((2)))

w1p = tf.Variable(tf.random_normal((2,2)))
b1p = tf.Variable(tf.zeros((2)))

w2 = tf.Variable(tf.random_normal((2,1)))
b2 = tf.Variable(tf.zeros((1)))
```

```
z1 = tf.matmul(x_,w1) +b1
a1 = tf.sigmoid(z1)
z1p = tf.matmul(a1,w1p) +b1p
a1p = tf.sigmoid(z1p)
z2 = tf.matmul(a1p,w2) + b2
a2 = tf.sigmoid(z2)
```

```

loss = -((y_* tf.log(a2)) +((1.0-y_)*tf.log(1.0-a2))) 
cost = tf.reduce_mean(loss) 
```

```
learning_rate = 0.1
op_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

```
XOR_X = [[0,0],[0,1],[1,0],[1,1]]
XOR_Y = [[0],[1],[1],[0]]
```

```
init = tf.global_variables_initializer()
sess = tf.Session()

sess.run(init)

for i in range(100000):
  sess.run(op_step, feed_dict={x_:XOR_X, y_:XOR_Y})

print('Final Prediction', sess.run(a2, feed_dict={x_: XOR_X, y_: XOR_Y}))
print('Final rounded Prediction', np.round(sess.run(a2, feed_dict={x_: XOR_X, y_: XOR_Y})))
```
