```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
def fct(x):
  return tf.square(x-0.5)
```

```
t = tf.placeholder(tf.float64, shape=(None), name='time')
B = fct(t)
```

```
n=1000.
step = 1./n
rng = np.arange(0., 1., step )
ts = rng.reshape((int(n),1))
```

```
sess = tf.Session()
```

```
Bs = sess.run(B, feed_dict={t:ts} )
plt.plot(ts, Bs)
```

```
learning_rate = 0.01
n_iter = 10000
```

```
grad = tf.gradients(B, t)
the_t = 0
```

```

for _ in range(n_iter):
  the_t = the_t -(learning_rate *sess.run(grad, feed_dict={t:the_t})[0])
```

```
print(the_t)
```
