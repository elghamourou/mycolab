```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
def f(x):
  return tf.square(x-0.5)

def g(x):
  return tf.cos(x)
```

```
v = tf.placeholder(tf.float64, shape=(None), name='v')
```

```
A = f(g(v))

W = g(v)
H = f(W)

D_A_v = tf.gradients(A, v)

D_W_v = tf.gradients(W,v)
D_H_W = tf.gradients(H,W)

product = D_W_v[0] * D_H_W[0]

```

```
n=10000.
lmin = -10.
umax = 10.
step = 1./n
rng = np.arange(lmin, umax, step )
vs = rng.reshape((int(n * (umax - lmin)),1))
```

```
sess = tf.Session()
```

```
result = sess.run([D_A_v, product, A], feed_dict={v:vs})
```

```
plt.plot(vs, result[2])
```

```
plt.plot(vs, result[0][0])
```

```
plt.plot(vs, result[1])
```

```
plt.plot(vs, result[1] - result[0][0])
```
